```python
# Drop rows where the 'State' column has specific values
suicide_df = suicide_df.drop(suicide_df[suicide_df['State'].isin(['Total (All India)', 'Total (States)', 'Total (Uts)'])].index)

# Verify the rows have been dropped
print(suicide_df['State'].unique())
```

```python
# Creating a series of total suicides according to year
yearwise_total_suicide = suicide_df.groupby("Year")["Total"].sum()
print('Total number of suicides in each year: ', yearwise_total_suicide)

# Plotting this in a line graph
import matplotlib.pyplot as plt

yearwise_total_suicide.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Total Suicides')
plt.title('Total Suicides per Year')
plt.show()
```

```python
#Creating a series of total suicides on the basis of states
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into suicide_df
suicide_df = pd.read_csv("C:/Users/IFC-48/Downloads/tech/Suicides in India 2001-2012 (1).csv")

# Creating a series of total suicides on the basis of states
statewise_total_suicide = suicide_df.groupby("State")["Total"].sum()
print('Total number of suicides in each State: ', statewise_total_suicide)

# Plotting the above data
statewise_total_suicide.plot(kind='bar', figsize=(12, 6))
plt.xlabel('State')
plt.ylabel('Total Suicides')
plt.title('Total Suicides per State')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into suicide_df
suicide_df = pd.read_csv("C:/Users/IFC-48/Downloads/tech/Suicides in India 2001-2012 (1).csv")

# Print the column names to ensure they match
print(suicide_df.columns)

# Assuming the correct column names are 'Year', 'Type_code', and 'Total'
# Adjust the column names if they are different
# Grouping the data by 'Year' and 'Type_code' and summing the 'Total' suicides
year_cause_suicide = suicide_df.groupby(['Year', 'Type_code'])['Total'].sum().unstack()

# Plotting the data
year_cause_suicide.plot(kind='bar', stacked=True, figsize=(14, 8))
plt.xlabel('Year')
plt.ylabel('Total Suicides')
plt.title('Suicides According to the Causes in Each Year')
plt.legend(title='Causes', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
