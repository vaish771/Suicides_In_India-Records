```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = "C:/Users/IFC-48/Downloads/cars.csv"
data = pd.read_csv(file_path)

# Strip leading and trailing spaces from column names
data.columns = data.columns.str.strip()

# Convert columns to numeric if they are not already, and handle non-numeric values
data['cubicinches'] = pd.to_numeric(data['cubicinches'], errors='coerce')
data['mpg'] = pd.to_numeric(data['mpg'], errors='coerce')
data['hp'] = pd.to_numeric(data['hp'], errors='coerce')
data['brand'] = data['brand'].str.strip()

# Drop rows with NaN values after conversion
data = data.dropna(subset=['cubicinches', 'mpg', 'hp', 'brand'])

# Calculate the correlation between cubic inches and mpg
ci_mpg_correlation = data['cubicinches'].corr(data['mpg'])
print('Correlation between cubic inches and mpg:', ci_mpg_correlation)

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['cubicinches'], data['mpg'])
plt.xlabel('Cubic Inches')
plt.ylabel('MPG')
plt.title('Scatter Plot of Cubic Inches vs. MPG')
plt.grid(True)
plt.show()

# Group by brand and calculate mean horsepower
hp_by_brand = data.groupby('brand')['hp'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
hp_by_brand.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Average Horsepower')
plt.title('Average Horsepower by Country')
plt.grid(True)
plt.show()
```

```python
data['cylinders'] = pd.to_numeric(data['cylinders'], errors='coerce')

# Group by number of cylinders and calculate mean mpg
mpg_by_cylinders = data.groupby('cylinders')['mpg'].mean()
print(mpg_by_cylinders)

# Create bar plot
plt.figure(figsize=(10, 6))
mpg_by_cylinders.plot(kind='bar')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.title('Average MPG by Number of Cylinders')
plt.show()
```

```python
correlation_matrix = data[['cubicinches', 'cylinders', 'hp']].corr()
print(correlation_matrix)

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix: Cubic Inches, Number of Cylinders, and Horsepower')
plt.show()
```

```python
 data.columns = data.columns.str.strip()  # Strip leading/trailing spaces from column names

# Convert cylinders column to numeric
data['cylinders'] = pd.to_numeric(data['cylinders'], errors='coerce')

# Calculate proportion of vehicles with 8 cylinders
total_records = len(data)
eight_cylinders_count = data[data['cylinders'] == 8].shape[0]
proportion = (eight_cylinders_count / total_records) * 100

print(f'Proportion of vehicles with 8 cylinders: {proportion:.2f}%')

# Create a pie chart to visualize the proportion
labels = ['8 Cylinders', 'Other']
sizes = [eight_cylinders_count, total_records - eight_cylinders_count]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # explode 1st slice

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Proportion of Vehicles with 8 Cylinders')
plt.show()
```
